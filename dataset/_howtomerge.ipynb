```python
from pandas import read_csv, DataFrame

patients = read_csv("patients.csv")
obs_before = read_csv("observations_before.csv")
obs_2h = read_csv("observations_2h.csv")
obs_5d = read_csv("observations_5d.csv")

merged_before = patients.merge(obs_before)
merged_before
```

```python
merged_two = merged_before.merge(obs_2h, on='id', suffixes=['_0', '_2h'])
merged_two
```

```python
obs_5d_s = obs_5d.add_suffix("_5d")
obs_5d_s = obs_5d_s.rename(columns={'id_5d': 'id'})
merged_three = merged_two.merge(obs_5d_s, on='id')
merged_three
```
